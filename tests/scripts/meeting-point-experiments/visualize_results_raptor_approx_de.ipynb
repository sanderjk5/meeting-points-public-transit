```python
results_avg_query_times = results[avg_query_times_results_columns].copy()
# remove the last column
results_avg_query_times = results_avg_query_times.iloc[:, :-1]
# rename the columns one by one
results_avg_query_times.columns = ["Naiver RAPTOR", "RAPTOR SELECT MinSum", "RAPTOR SELECT MinMax", "RAPTOR SELECT CAND MinSum", "RAPTOR SELECT CAND MinMax", "RAPTOR SELECT LOOP"]

# plot average query times
plt.figure()
results_avg_query_times.plot()
plt.title("Average Query Times")
plt.xlabel("Number of Source Stops")
plt.ylabel("Time (ms)")
plt.grid()
plt.show()
```
