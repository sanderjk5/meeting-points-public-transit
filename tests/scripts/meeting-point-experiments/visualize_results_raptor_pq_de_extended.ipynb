# Visualize Test Results (Raptor PQ) - Germany Graph

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Load the data

```python
folder = "../../gesamt_de/results/"

results_0 = pd.read_csv(folder + "raptor-pq-results-2024-7-22-8-49-4.csv") # k = 2, 3, 5
results_1 = pd.read_csv(folder + "raptor-pq-results-2024-7-19-16-40-8.csv") # k = 7
results_2 = pd.read_csv(folder + "raptor-pq-results-2024-7-20-11-31-15.csv") # k = 9
results_3 = pd.read_csv(folder + "raptor-pq-results-2024-7-20-16-8-37.csv") # k = 10

# # combine results
results = pd.concat([results_0, results_1, results_2, results_3])

# set number of source stops as index
results = results.set_index("numberOfSourceStops")

# sort by index
results = results.sort_index()
```

```python
display(results)
```

## Analyze the query times

```python
avg_query_times_results_columns = [col for col in results.columns if 'avgQueryTime' in col and not 'Parallel' in col]
median_query_times_results_columns = [col for col in results.columns if 'medianQueryTime' in col and not 'Parallel' in col]
max_query_times_results_columns = [col for col in results.columns if 'maxQueryTime' in col and not 'Parallel' in col]
min_query_times_results_columns = [col for col in results.columns if 'minQueryTime' in col and not 'Parallel'in col]
```

```python
display(results[avg_query_times_results_columns])
```

```python
display(results[median_query_times_results_columns])
```

```python
display(results[max_query_times_results_columns])
```

```python
display(results[min_query_times_results_columns])
```

```python
# copy the avg_query_times_results_columns
results_avg_query_times = results[avg_query_times_results_columns].copy()
# rename the columns one by one
results_avg_query_times.columns = ["RAPTOR1", "Naiver RAPTOR", "RAPTOR PQ MinSum", "RAPTOR PQ MinMax", "RAPTOR B MinSum", "RAPTOR B MinMax"]

# plot average query times
plt.figure()
# plot all columns but only the first one as dotted line
results_avg_query_times.iloc[:, 0].plot(ax=plt.gca(), linestyle='dotted')
results_avg_query_times.iloc[:, 1:].plot(ax=plt.gca())
# the legend should contain the labels of all columns
plt.legend(results_avg_query_times.columns)
plt.title("Durchschnittliche AusfÃ¼hrungszeit")
plt.xlabel("# Ausgangsstopps")
plt.ylabel("AusfÃ¼hrungszeit (in ms)")
plt.grid()
plt.show()
```
